```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('nbs_data.xlsx')
df_y_a = df.loc[:, ['Year', 'Abstract','Index Keywords']]

# 요약문에서 실제로 내용이 있는 부분('@' 앞)만을 가져와, 새 컬럼 Abs_split에 저장
abs_split = df_y_a['Abstract'].str.split('©')
df_y_a['Abstract'] = abs_split.str.get(0)
df_y_a


if (df_y_a["Index Keywords"].isnull().any().any()):
    df_y_a["abs+key"] = df_y_a["Abstract"]
else:
    df_y_a["abs+key"] = df_y_a["Abstract"] + " "+ df_y_a["Index Keywords"]
    
df_y_a
```

```python
df_y_a['abs+key'][21]
```

```python
#정렬하기 
df_sorted_by_values = df_y_a.sort_values(by='Year',ascending=True)
df_sorted_by_values.head(1056)
```

```python
df_all_list = df_y_a.groupby(['Year']).count()
df_all_list = sum(df_all_list.values.tolist(), [])
df_all_list
```

```python
year = [2003, 2006, 2007, 2009, 2010, 2012, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
keyword = ["forest", "forestation", "afforestation", "reforestation"]
tree_keyword = ["forest", "forestation", "afforestation", "reforestation", "tree"]
plant_keyword = ["forest", "forestation", "afforestation", "reforestation", "plant"]
tree_plant_keyword = ["forest", "forestation", "afforestation", "reforestation", "tree", "plant"]
```

```python
df_list = df_y_a.values.tolist()
df_pd= pd.DataFrame(df_list)
df_pd.columns=['Year','Abstract','Index Keywords','Abs+key']
df_pd = df_pd.loc[:, ['Year', 'Abs+key']]
```

```python
#연도별 keyword 개수
df_pd1=df_pd[df_pd['Abs+key'].str.contains('|'.join(keyword))]
df_keyword = df_pd1.groupby(['Year']).count()
df_keyword
```

```python
#연도별 tree_keyword 개수
df_pd2=df_pd[df_pd['Abs+key'].str.contains('|'.join(tree_keyword))]
df_tree_keyword = df_pd2.groupby(['Year']).count()
df_tree_keyword
```

```python
#연도별 plant_keyword 개수
df_pd3=df_pd[df_pd['Abs+key'].str.contains('|'.join(plant_keyword))]
df_plant_keyword = df_pd3.groupby(['Year']).count()
df_plant_keyword
```

```python
#연도별 tree_plant_keyword 개수
df_pd4=df_pd[df_pd['Abs+key'].str.contains('|'.join(tree_plant_keyword))]
df_tree_plant_keyword = df_pd4.groupby(['Year']).count()
df_tree_plant_keyword
```
